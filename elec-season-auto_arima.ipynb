```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pmdarima.arima import auto_arima
from pmdarima.arima import ndiffs
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.stattools import adfuller
```

```python
data = pd.read_csv('data/electricity.csv')
data
```

```python
df = data[['dates', 'Total']] #dates total df 

df = df.rename(columns={
    "Total" : "total"
}) # Total  

df
```

```python
df.info() #type  null 
```

```python
df[df['total'].isnull()]
```

```python
df['dates'] = pd.to_datetime(df['dates']) #dates   
df.info()
```

```python
df.set_index('dates',inplace=False)
```

```python
df_all = df[['total']]

df_all.plot()
```

```python
df_all_null = df_all.dropna(axis=0) # NAN plot . 
# df_all_null.info()
plot_acf(df_all_null['total'], lags=30, alpha=0.05)
```

```python
df['year'] = df['dates'].dt.year # dates  year 
df
```

```python
df9703 = df.query('year >= 1997 & year <= 2003') # 1997-2003  
df9703
```

```python
df9703 = df9703[['total']]
df9703
```

```python
df9703.plot()
```

```python
df9703.describe()
```

```python
plot_acf(df9703['total'], lags=30, alpha=0.05)
```

```python
plot_pacf(df9703['total'], lags=30, alpha=0.05, method='ywm')
```

```python
def adf_test(x):
    stat, p_value, lags, nobs, crit, icb = adfuller(x)
    
    return p_value

print('pvalue =', adf_test(df9703))
```

```python
kpss_diffs = ndiffs(df9703, alpha=0.05, test='kpss', max_d=6)
adf_diffs = ndiffs(df9703, alpha=0.05, test='adf', max_d=6)
n_diffs = max(adf_diffs, kpss_diffs)

print(f" d = {n_diffs}")
```

```python
df9703_arima = auto_arima(y = df9703        # 
                      , d = 1            #  , ndiffs !
                      , start_p = 0 
                      , max_p = 3   
                      , start_q = 0 
                      , max_q = 3   
                      , m = 12      
                      , seasonal = True #  ARIMA  !
                      , stepwise = True
                      , trace=True      # stepwise     
                      )
```

```python
# ..

auto_str = str(df9703_arima)

print(auto_str)
```

```python
df9703_arima.fit(df9703)
```

```python
print(df9703_arima.summary()) #   1
```

```python
df9703_arima.plot_diagnostics(figsize=(16, 8)) #   1
plt.show()

#     : Standardized residual Correlogram
#    : Histogram plus estimated density Normal Q-Q

# Standardized residual     
# Correlogram   ACF
# Histogram plus estimated density      N(0,1)   
# Normal Q-Q Q-Q        

## -->   . 
```

```python
df9703_fit = df9703_arima.fit(df9703)
print(df9703_fit.summary())
```

```python
y_predict = df9703_fit.predict(start=df9703.index[0],end=df9703.index[-1], typ='levels')
# y_predict = pd.DataFrame(y_predict,index = df9703.index,columns=['Prediction'])
# y_predict = df9703_fit.predict(n_periods=len(df))

# 
plt.plot(df9703, label='Original data')
plt.plot(y_predict, label='ARIMA model')

plt.legend()
plt.show()
```
